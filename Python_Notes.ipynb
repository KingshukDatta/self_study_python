# Global Variables
Variables that are created outside of a function are known as global variables.

Global variables can be used by everyone, both inside of functions and outside.

If we create a variable with the same name inside a function, this variable will be local, and can only be used inside the function. The global variable with the same name will remain as it was, global and with the original value.

````python
x = "awesome" # Global Variable

def myfunc():
  x = "fantastic" # Local Variable
  print("Python is " + x)

myfunc()

print("Python is " + x)
````

# The global Keyword
Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function.

To create a global variable inside a function, you can use the global keyword.

````python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
````

A **frozenset** in Python is essentially a "locked" version of a regular set.

Just like a tuple is an immutable (unchangeable) version of a list, a frozenset is an immutable version of a set.

Key Characteristics
* Immutable: Once you create it, you cannot add, remove, or update elements.

* Unique Elements: Like a normal set, it removes duplicates automatically.

* Unordered: The items have no fixed index or order.

* Hashable: Because it never changes, Python can assign it a unique "hash" value. This is the most important feature.



````python
# Creating a normal set
normal_set = {1, 2, 3}

# Creating a frozenset
frozen = frozenset([1, 2, 3, 3, 2])

print(frozen)
# Output: frozenset({1, 2, 3})  <-- Duplicates removed automatically
````

Why do we need it? (The "Hashable" Logic)
This is a core Computer Science concept.

* Rule: You cannot use "mutable" items (like lists or normal sets) as keys in a dictionary, because if the item changes later, the dictionary breaks.

* Solution: Since frozenset is immutable, you can use it as a dictionary key.

````python
# ❌ THIS WILL FAIL (TypeError)
# locations = {
#     {"Lat", "Long"}: "New York"
# }

# ✅ THIS WORKS
locations = {
    frozenset({"Lat", "Long"}): "New York",
    frozenset({"X", "Y"}): "Graph Point"
}

print(locations[frozenset({"Lat", "Long"})])
# Output: New York
````

# **Note:**
we can not convert Complex to Int or float because, colplex datatype has 2 parts, real and complex.

Python does not know which part you want to keep:

1.   Do you want the imaginary number?
2.   Do you want the magnitude (length) of the number?
3.   Do you want just the real number?

Because this is ambiguous, Python refuses to guess and raises an error instead.

```
z = 3 + 4j
# print(int(z))  
# ❌ ERROR: TypeError: can't convert complex to int
```
But we can convert by telling which part is needed to be changed.


# Converting only the Real part
You can access the real part using .real, which gives you a float. You can then cast that to an int.

````python
z = 3 + 4j
real_part = int(z.real)
print(real_part)
# Output: 3
````

# Converting only the Imaginary part
You can access the imaginary part using .imag.

````python
z = 3 + 4j
imag_part = int(z.imag)
print(imag_part)
# Output: 4
````

# Converting the Magnitude (Absolute Value)
In fields like Signal Processing or Electrical Engineering (very relevant to VLSI), you often care about the "magnitude" (or strength) of the complex signal.$$|z| = \sqrt{a^2 + b^2}$$

````python
z = 3 + 4j
magnitude = abs(z)      # Result is 5.0 (Float)
final_int = int(magnitude)
print(final_int)
# Output: 5
````

# F-Strings
F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.

````python
age = 36
txt = f"My name is John, I am {age}"
print(txt)
````

A placeholder can include a modifier to format the value.

A modifier is included by adding a colon : followed by a legal formatting type, like .2f which means fixed point number with 2 decimals:

````python
price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
````

A placeholder can contain Python code, like math operations:

````python
txt = f"The price is {20 * 59} dollars"
print(txt)
````

Almost any value is evaluated to True if it has some sort of content.

Any string is True, except empty strings.

Any number is True, except 0.

Any list, tuple, set, and dictionary are True, except empty ones.
